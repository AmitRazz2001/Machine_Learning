```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_breast_cancer
```

```python
df = pd.DataFrame(cancer.data,columns = cancer.feature_names)
df
```

```python
df["cancer"] = cancer.target
df.head()
```

```python
df = df.drop("Target", axis=1)
df
```

```python
scaler = StandardScaler()
scaled = scaler.fit_transform(df)
scaled
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
yp = pca.fit_transform(scaled)
yp
```

```python
plt.scatter(yp[:,0],yp[:,1],c = cancer.target)
```

```python
df1 = pd.DataFrame(yp,columns=["X","Y"])
# df1["cancer"] = cancer.target
df1
```

```python
from sklearn.cluster import KMeans

wcss = []
for i in range(1,10):
    km = KMeans(n_clusters=i)
    km.fit(df1)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,10),wcss)
plt.xlabel("K")
plt.ylabel("WCSS")
```

```python
km = KMeans(n_clusters=2)
km.fit_predict(df1)
```

```python
df1["clusters"] = km.fit_predict(df1)
df1.head()
```

```python
plt.scatter(df1.X,df1.Y,c = km.fit_predict(df1),cmap = "rainbow")
```

```python
plt.scatter(yp[:,0],yp[:,1],c = cancer.target,cmap = "rainbow")
```

```python
x = df1.drop("clusters",axis = 1)
y = df1.clusters
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=2)
model.fit(x_train,y_train)
```

```python
pred = model.predict(x_test)
pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report 
```

```python
cm = confusion_matrix(y_test,pred)
cm
```

```python
print(accuracy_score(y_test,pred))
```

```python
model.score(x_train,y_train)
```

```python
print(classification_report(y_test,pred))
```
